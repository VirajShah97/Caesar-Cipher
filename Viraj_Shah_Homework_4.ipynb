### Step 3.1: Scrape the main image for Bulbasaur in a general way that could be applied to other pokemon pages by searching for the relevant tag and extracting the image URL.

```python
def image_scrape(imgsrc):
    url = 'https://pokemondb.net/pokedex/' + imgsrc
    content = requests.get(url).content
    soup1 = BeautifulSoup(content,'html.parser')
    image_tags = soup1.find_all('img', {'src':re.compile('.jpg')})
    for image_tag in image_tags:
        print("THE IMAGE URL IS : {}" .format(image_tag.get('src')))
```

```python
image_scrape('bulbasaur')
```
